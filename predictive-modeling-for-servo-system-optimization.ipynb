# Servo Motor Mechanism Dataset

## Description:
This dataset provides information about a servo motor mechanism, capturing various parameters and their corresponding classes. Each entry in the dataset represents a specific configuration of the servo motor mechanism along with associated attributes.
  
![](https://www.googleapis.com/download/storage/v1/b/kaggle-user-content/o/inbox%2F19517213%2F76e6bdc47f6b0459da63e1508bdd2344%2F12d8ac11ee020eefc1f76900aa7a271c.jpg?generation=1712418501189517&alt=media)

## Attributes:

- **Motor:** Represents the type of motor used in the mechanism, categorized as 'A', 'B', 'C', 'D', or 'E'.
- **Screw:** Indicates the type of screw used in the mechanism, classified as 'A', 'B', 'C', 'D', or 'E'.
- **Pgain:** Refers to the proportional gain setting for the servo motor.
- **Vgain:** Represents the velocity gain setting for the servo motor.
- **Class:** Denotes the class label assigned to each configuration, which is a measure related to the performance or behavior of the servo mechanism.

## Dataset Information:
The dataset contains a total of 5 entries, each representing a unique configuration of the servo motor mechanism. This dataset can be used for tasks such as classification or regression to predict the class label or performance metric based on the given attributes.

## Potential Uses:

Servo Mechanism Performance Prediction: Build machine learning models to predict the performance class of a servo mechanism based on its attributes.
Feature Engineering and Selection: Explore the relationships between different attributes ( Motor type, Screw type, Pgain, Vgain) and their impact on servo mechanism performance.
Educational Purposes: Use this dataset for educational activities such as data analysis, visualization, and modeling within the context of servo motor mechanisms.

## Data Source:
This dataset was originally sourced from a study or experiment involving servo motor mechanisms and has been made available for analysis and exploration. The dataset is suitable for use in machine learning projects, educational settings, and research related to servo motor systems and control mechanisms.

```python
# Unique values in Class Column
servo['Class'].nunique()
```

```python
# Class Column Counts 

servo['Class'].value_counts().sort_index()
```

```python
# creatu=ing range for Class column by Rise time of system
servo['RiseTime_Category'] = None  # Initialize the new column

# Define the categories
categories = {
    range(1, 11): 'Very Fast',
    range(11, 21): 'Fast',
    range(21, 31): 'Medium',
    range(31, 41): 'Slow',
    range(41, 52): 'Very Slow'
}

for i in range(len(servo)):
    for category_range, category_label in categories.items():
        if servo.loc[i, 'Class'] in category_range:
            servo.loc[i, 'RiseTime_Category'] = category_label
            break
```

```python
servo[servo['RiseTime_Category']== 'Very Slow'].value_counts().sum()
```

# **Step - 4 : Visualize the Data**


**4.1Barplot : Class Dist. by Motor type and Screw type**
--

```python
# Use Group by function 
plt.figure(figsize=(8,6))

motor_screw_counts = servo.groupby(['Motor','Screw'])['Class'].sum().reset_index()
sns.barplot(x='Motor',y='Class',hue='Screw',
           data=motor_screw_counts,
           palette='coolwarm',edgecolor='black')
plt.title("Class distribution by Motor type and Screw type")
plt.show();
```

Insights:
--

- Screw type (Screw) selection significantly influences Class values for each motor type (Motor).

- Motor C with Screw A (C-A) exhibits the highest Class value (251), suggesting an optimal combination.

- Comparisons across motor-screw combinations reveal performance variations and potential areas for improvement.

# **Step - 5 : Label Encoding on Categorical Data**
<hr>

```python
## Ordinal encoding on Categorical data
servo.replace({'Motor':{'A':0,'B':1,'C':2,'D':3,'E':4}},inplace=True) #ordinal encoding
servo.replace({'Screw':{'A':0,'B':1,'C':2,'D':3,'E':4}},inplace=True) #ordinal encoding
servo.replace({'RiseTime_Category':{'Very Fast':1, 'Fast':2, 'Medium':3, 'Slow':4, 'Very Slow':5}},inplace=True) #ordinal encoding
```

```python
servo.head()
```

# **Step - 6 : Split the Data**
<hr>

```python
X = servo.drop(columns=['RiseTime_Category']) #features
y = servo['RiseTime_Category'] #target variable
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
```

# **Logistic Regression**
## **Model Implementation**
-----------------

```python
from sklearn.linear_model import LogisticRegression
logit=LogisticRegression(multi_class = 'ovr')
```

```python
logit.fit(X_train,y_train)
```

```python
# Predictions on Testing Data
y_pred_lr = logit.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
a_score = accuracy_score(y_test,y_pred_lr)
print("Accuracy of Logistic Regression = %.2f"%(a_score*100),'%')
print("Percentage Error of Logistic Regression = %.2f"%((1-a_score)*100),'%')
```

------------------------------

# **Decision Tree Classifier**
## **Model Implementation**
-----------------

```python
from sklearn.tree import DecisionTreeClassifier
# Specify hyperparameters using keyword arguments
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(X_train,y_train)
```

Traning Predictions 
--
------------

```python
y_pred_t_dtc = dtc.predict(X_train)
```

```python

a_score = accuracy_score(y_train,y_pred_t_dtc)
print("Accuracy of Decision Tree Classifier = %.2f"%(a_score*100),'%')
print("Percentage Error of Decision Tree Classifier = %.2f"%((1-a_score)*100),'%')
```

Testing Predictions 
--
----------

```python
y_pred_dtc = dtc.predict(X_test)
```

```python
a_score = accuracy_score(y_test,y_pred_dtc)
print("Accuracy of Decision Tree Classifier = %.2f"%(a_score*100),'%')
print("Percentage Error of Decision Tree Classifier = %.2f"%((1-a_score)*100),'%')
```

```python
dtc.predict(X.sample(1))
```

Insights and Observations:
--
- **High Test Accuracy:**

  A test accuracy of 97.06% indicates that the Decision Tree Classifier, after hyperparameter tuning, can generalize effectively to new, unseen data.
  The model demonstrates strong predictive performance in classifying instances based on the learned patterns from the training data.
- **Generalization Capability:**

  Achieving a high test accuracy suggests that the model is not overfitting to the training data.
  The tuned hyperparameters likely contribute to a well-balanced model that captures underlying patterns without memorizing noise from the training set.
- **Model Robustness:**

  The test accuracy provides confidence in the model's ability to make accurate predictions in real-world scenarios.
  The tuned Decision Tree Classifier shows robustness in handling variability and complexity present in unseen data.
